```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
df.dropna(subset=['Amount'], inplace=True)

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Age Group', y='Amount', data=sales_age, palette='Set2')

# Add titles and labels for clarity
plt.title('Total Sales Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.savefig('sales_by_age_group_color.png')

print(sales_age)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

ax = sns.barplot(data = sales_state, x = 'State', y = 'Orders', palette='Set2')

for container in ax.containers:
    ax.bar_label(container)

plt.title('Top 10 States by Total Orders', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Total Orders', fontsize=14)
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()

plt.savefig('top_10_states_orders_color.png')
#plt.show() 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

df.dropna(how='all', inplace=True)

occupation_counts = df['Occupation'].value_counts().index

sns.set(rc={'figure.figsize':(20,6)})

ax = sns.countplot(data = df, x = 'Occupation', order = occupation_counts, palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Count of Customers by Occupation (Sorted)', fontsize=16)
plt.xlabel('Occupation', fontsize=14)
plt.ylabel('Count of Customers', fontsize=14)
plt.xticks(rotation=45, ha='right')

plt.savefig('occupation_countplot.png')
```

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

df_cleaned = df.dropna(subset=['Amount'])

top_5_states_list = df_cleaned.groupby('State')['Amount'].sum().nlargest(5).index.tolist()

df_top_states = df_cleaned[df_cleaned['State'].isin(top_5_states_list)]

top_5_categories_list = df_top_states.groupby('Product_Category')['Amount'].sum().nlargest(5).index.tolist()

df_final = df_top_states[df_top_states['Product_Category'].isin(top_5_categories_list)]

sales_data = df_final.groupby(['State', 'Product_Category'])['Amount'].sum().reset_index()

sales_data['Amount_Cr'] = sales_data['Amount'] / 10**7

sns.set(style="whitegrid")
plt.figure(figsize=(18, 8))

ax = sns.barplot(
    data=sales_data,
    x='State',
    y='Amount_Cr',
    hue='Product_Category', 
    palette='Spectral' 
)

plt.title('Top 5 Product Categories Sales in Top 5 States (Amount in Cr.)', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Total Sales Amount (in Cr.)', fontsize=14)
plt.xticks(rotation=0)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', fontsize=8)

plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.savefig('top5_states_top5_categories_sales.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

df.dropna(subset=['Amount'], inplace=True)

sales_by_category = df.groupby('Product_Category')['Amount'].sum().reset_index()

sales_by_category = sales_by_category.sort_values(by='Amount', ascending=False)

plt.figure(figsize=(14, 7))

sns.barplot(
    x='Product_Category',
    y='Amount',
    data=sales_by_category,
    palette='viridis'
)

plt.title('Total Sales Amount by Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Sales Amount (in Lakhs)', fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.ticklabel_format(style='plain', axis='y') 

plt.tight_layout()

plt.savefig('total_sales_amount_by_product_category.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

df_clean = df.drop(['Status', 'unnamed1'], axis=1)

df_clean.dropna(inplace=True)

orders_distribution = df_clean.groupby(['Age Group', 'Gender'])['Orders'].sum().reset_index()

plt.figure(figsize=(10, 6))

sns.barplot(data=orders_distribution, x='Age Group', y='Orders', hue='Gender', palette='magma')

plt.title('Orders Distribution by Age Group and Gender', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Total Orders', fontsize=14)

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig('orders_distribution_age_gender.png')
```

### 1). Target Customers by Gender: Focus marketing and special offers primarily on Female customers, as they are the main buyers and contribute the most to the total sales amount.

### 2). Focus on Age Group: Direct sales and advertising efforts towards the 26-35 age group, as this group has the highest purchasing power and makes the most purchases.

### 4). Target Marital Status: Design exclusive deals for Married customers, as they contribute more to the overall sales amount compared to unmarried customers.

### 5). Promote Top Categories: Focus inventory and promotions on the highest-selling categories: Food, Clothing & Apparel, and Electronics & Gadgets. These categories should be given maximum visibility and discounts.

### 6). Targeted Marketing by Occupation: Consider running targeted campaigns for people working in the IT Sector, Healthcare, and Aviation professions, as these groups show high purchase activity (implied from the customer count analysis).
